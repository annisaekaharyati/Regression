Here, I present a simple and easy way to create a multiple linear regression.
The file multiple_linear_regression.csv contains generic data, with the only purpose of example.


# The Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import statsmodels.api as sm
```

# The data 

```python
data = pd.read_csv('multiple_linear_regression.csv')
data.head()
```

Verifying that data is completed

```python
data[pd.isnull(data['OPTION'])]
```

learning about this data set

```python
data.describe()
```

# The Regression

```python
# The dependent variable (y) is the DATA
y = data ['DATA']
# The independent variableS (x) are the YEAR and the option
x_1 = data[ ['YEAR','OPTION']]

#Plot this data
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl

fig = plt.figure()
ax = plt.axes(projection='3d')

# three-dimensional scattered points
ax.scatter3D(x_1.YEAR, x_1.OPTION, y);
ax.set_xlabel('YEAR')
ax.set_ylabel('OPTION')
ax.set_zlabel('DATA')
```

```python
#The regression model

# Add a constant. Esentially, we are adding a new column (equal in lenght to x), which consists only of 1s
x = sm.add_constant(x_1)
# Fit the model, according to the OLS (ordinary least squares) method with a dependent variable y and an idependent x
Result = sm.OLS(y,x).fit()
#analyze the model
Result.summary()
```

# The results

```python
# The same plot

fig = plt.figure()
ax = plt.axes(projection='3d')

# The regression equation,from previous model
y_hat = 0.0006*x_1.YEAR - 0.0022*x_1.OPTION + 2.0383
ax.scatter3D(x_1.YEAR, x_1.OPTION, y_hat, c=y_hat, cmap='Greens');


# three-dimensional scattered points
ax.scatter3D(x_1.YEAR, x_1.OPTION, y);
ax.set_xlabel('YEAR')
ax.set_ylabel('OPTION')
ax.set_zlabel('DATA')
```
